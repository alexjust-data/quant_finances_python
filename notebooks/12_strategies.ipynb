**Long Position**  
>Si un inversionista tiene posiciones largas, significa que el inversionista ha comprado y posee esas acciones. Por el contrario, si el inversor tiene posiciones cortas, significa que el inversor le debe esas acciones a alguien, pero en realidad aún no las posee.
>
>Por ejemplo, se dice que un inversor que posee 100 acciones de Tesla ( TSLA ) en su cartera tiene 100 acciones largas. Este inversionista ha pagado en su totalidad el costo de poseer las acciones.

**Short Position**  
>Continuando con el ejemplo, se dice que un inversor que ha vendido 100 acciones de TSLA sin poseer aún esas acciones tiene 100 acciones cortas. El inversor corto debe 100 acciones al momento de la liquidación y debe cumplir con la obligación comprando las acciones en el mercado para entregar.
>
>A menudo, el inversionista corto toma prestadas las acciones de una firma de corretaje en una cuenta de margen para realizar la entrega. Luego, con la esperanza de que el precio de las acciones caiga, el inversionista compra las acciones a un precio más bajo para pagarle al corredor que las prestó. Si el precio no baja y sigue subiendo, el vendedor en corto puede estar sujeto a una llamada de margen de su corredor.

## 3. ¿Qué es Central Risk Book (CRB)?

Desde el punto de vista de la diversificación de portafolio, mientras más activos tenga en un portafolio, la contribución de épsilon va a ser cada vez más pequeña [principio de diversificación de portafolios].  Esto significa que 
* si nos cronstruímos un portafolios lo suficientemente diversificado, épsilon lo podemos tirar a la basura, 
* si dicho portafolio es además beta-neutral significa que toda la ganacia o pérdida va a ser explicada por alpha del portafolio.

Entonces muchas veces tener un portafoio delta-neutral es porque 
* 1MM contra 1MM se le llama market neutral es la primera aproximación
* la segunda aproximación que es un poco más avanzada que requiere conocer CAPM es beta-neutral
* pero cuando son delta y beta neutral y diversificado, significa que la única componente que nos va a dar rendimiento es el alpha (al menos en toería) alpha puro.

**¿Qué es Central Risk Book?**  
Es algo que comenzó cuando fue la crisis de las subprimes porque se prohibioi durante esas crisis porque se estaba restringiendo tomar posiciones especulativas, pero había ciertas reglas de las cuales decían que si habían cosas que sí puedes tomar posiciones porque si estás ofreciendo iquidez a clientes, si estás tomando estrategias de cobertura, entonces si puedes tomar dichas posiciones.

* https://www.efinancialcareers.co.uk/news/2018/02/central-risk-desks-banking


Cuando tenemos un departamento y tenemos en el traders dedicados lo que hacen esos traders (están manegando un portafolio con un inventario dado) no tiene control en genral sonbre las cosas que entran en el portafolio. Puede ser porque clientes interactúan con nosotros, porque clientes buscan posiciones de liquidez o puede ser porque otros traders están haciendo delta-hedges, alpha-hedges o coberturas; no usan como fuente de liquidez y no teneos el control de quien y cuándo nos compran o nos venden productos, pero un vez que los productos entran en el portafolio de riesgo central dichos productos ya son parte nuestra y ahí tengo completa libertad de hacer lo que quiero, tengo libertad de quedarme con ella, de deshacerme de ella, cómo me deshago de ella, la cubro? o cómo la cubro. Ahora como imaginarán, si es un banco en el cual interactúas con todos los traders del banco más con clientes externos es imposible hacerlo de manera manual, has de crearte algoritmos que sigan ciertas reglas y busquen tres reglas en particular:
1. minimizar riesgos de transacción
> * Por ejemplo si en mi banco alguien vende vodafone y en mi banco alguien está comprando vodafone, entonces antes de que ambos vayan al mercado paguen costos de transacciones y tengan impacto de mercado, estaría más fácil que el que compra le compre al Central Risk Book el que venda se lo venda al Central Risk Book y se haga un neck de posiciones. 
> * Sólo iremos al mercado con las posiciones residuales despues de hacer el meeting

2. Optimize risk management of trading positions
> * Define risk (and the metrics of quantify it) Si contruimos el algoritmo de cómo manegar el riesgo de inventario esto significa que estamos construyendo una estrategia de gestión de riesgo optimizada y automatizada.
> * Determine the adequade risk budge. Osea que hemos de determinar el adecuado apetito de riesgo (risk budget) 
> * Cuando el presupuesto de riesgo es sufiencientemente grande y las posiciones actuales no superan esas psoiciones de riesgo no hacemos nada y no salimos al mercado. No hay costos de transacción, pero si lo sobrepasamos entonces hemos de salirnos de la posición o bien estrategias de cobertura.

3. Enhance liquidity offering
La razón principal por la que existe un CRB es para ofrecer liquidez.
> * Indications of Interest (IOIs): blocks. Si un cliente dice, quiero comprar 20MM de HSBC si va directo al mercado tendrá mucho impacto, pero si hace la transacción con CRB no lo hará; pero esto significa que si ellos quieren comprar ahora nosotros vamos a tener una posición en corta de 20MM porque nosotros vendemos y hay que saber qué hacemos, si nos salimos del mercado o la cubrimos.
> * Systematic INternaliser (SI): maret making. Pero también podemos funcionar como un mercado de liquidez, tanto internamente a otros traders como ha ciertos clientes. les decimos que estmoas dispuesto a comprar y a vender estos activos a este precio. Esto significa que de cierta manera funcionamos como market making.
