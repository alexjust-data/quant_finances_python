## Coberturas de portafolios DELTA neutral vs BETA neutral

Supongamos que me dan un activo (puede ser un portfolios) y le llamamos $S_0$:
* Let $S_0$ be its value in USD (p.e 1000000€ en la acción de Santander)

**COBERTURA CON UNA SEGURIDAD**

|**`Delta neutral`** |**`Beta neutral`** |
|-|-|
|Entonces, si yo quiero hacer una cobertura `Delta neutral` para Santander, esto significa que me quiero construir un *portafolios de mercado neutral* cuyo valor del Delta sea cero: Find $S_1$ such that $S_0 + S_1 = 0$ (si tengo 1MM de Santander, vendo 1MM de Bbva)| Tomamos 1MM del Santander $S_0$ con respecto al STOXX600, como el beta $\beta_0$ es supongamos 1.4 para STOXX600, entonces el riesgo en exposición de beta se transforma en 1.4MM. Luego tenemos otro beta $\beta_1$ de 1.3 para el Bbva, entonces para que sea neutral voy a tener que poner más de Bbva $S_1$ para que me dé el mismo valor y equilibrarlo a 0: Find $S_1$ such that $\beta_0 S_0 + \beta_1 S_1 = 0$|


Un portafolios **beta neutral** significa que este $\beta_0 S_0 + \beta_1 S_1 = 0$ es el beta total de todo mi portafolio porque mi portafolio aquí tiene dos activos, el beta de mi portafolio es 0. Ahora

**¿Cuál de las dos queremos?**  
Quizás queremos las dos a la vez:
* `Delta neutral` : que las compras y las ventas tengan el mismo nocional, es decir que tengo 10MM en compras y 10MM en ventas.
* Pero también quisiera que mi `Beta` de mi portfolio fuera neutral porque debido al Capital Asset Pricing Model (CAPM) si yo tengo un beta=0 significa que tengo una cobertura total del portafolio y además desde el punto de vista del mercado.
>
> ![4.png](img/04.png)
> Si mi beta=0 significa que sólo tengo dos exposiciones de riesgo, una es el $\alpha$ y el otro es el $\varepsilon$ pero si estoy jugando con la **hipótesis de mercados eficientes** significa que también el alpha es 0 y si además me tomo esperanza significa que el riesgo total es cero porque todo esto de aquí $E[r_a] = \beta E[r_M]$ va a ser cero porque beta es cero; tengo un activo que no se mueve, lo tengo totalmente cubierto. Es lo que busco aquí  $\beta_0 S_0 + \beta_1 S_1 = 0$ ; lo que le puse de Bbva junto con el Santander es una perfecta cobertura de portfolio.

**COBERTURA CON $N$ SEGURIDADES**
>Si tengo $N$ activos para cubrir
>
>**`portfolio Delta neutral`**: Find $S_1, ..., S_N$ such that $S_0+\sum S_n = 0$ 
>* $\sum S_n$ será exactamente menos que el valor original $S_0$
>
>**`portfolio Beta neutral`** : 
>* Find $S_1,...,S_N$ such that $\beta_0 S_0+\sum \beta_n S_n = 0$


**Una estrategia de cobertura depende de la métrica de riesgo que usamos y el modelo**  
¿Podríamos decir que la cobertura va en función de la medida de riesgo en este caso del CAPM?
Absolutamente, en el caso del CAPM mi medida de riesgo en el beta, por lo tanto yo estoy buscando un portfolio que tenga la menor candidad de riesgo desde esa perspectiva. Lo cual significa que tengo un portfolio beta=0.

Podemos hacer una cobertura de portfolio de diferente manera, por ejemplo en **opciones** tenemos un portfolio que tenemos una **opción** y quiero comprar el **activo**. Cuando tengo un *Call - Delta* veces el activo, este es un portfolio que tiene un delta=0 porque el Delta del Call es $\Delta S $ y estamos vendiendo $\Delta S $ entonces esto es cero.

Incluso podemos decir que queremos una cobertura que sea **Gamma** neutral o **Vega** neutral, entonces significa que yo voy a agreagr a mi portfolio un cierto numero de activos tal que las métricas que tomé sean igual a cero. Esta es una estrategia de cobertura.

En nuestro caso como el modelo es CAPM y estoy suponeniendo la teoría de **mercados eficientes**, todo se reduce al beta porque el epsilon se va cuando tomamos esperanzas.

## Beta and delta neutral hyperplanes

**Delta-neutral hyperplane in $R^N$**  
En la dimensionalidad R el $N$ es el número de activos. Entonces cuando consideramos 
* $L_{delta} = \{S_0 + \sum S_n = 0 \}$ entonces tenemos $N$ activos contando mi activo original que es $S_0$ o sea que la sumatoria es n-1.
* Yo estoy buscando coberturas, entonces el hiperplano $L_{delta}$ neutral son todos los activos que satisfacen $\{S_0 + \sum S_n = 0 \}$. Todos esos puntos me dan delta-neutral.

**Delta-neutral hyperplane in $R^N$**  
Exactamente lo mismo pero le cambio los pesos.
* $L_{delta} = \{\beta_0 S_0 + \sum \beta_n S_n = 0 \}$
* Yo estoy buscando coberturas, entonces el hiperplano $L_{delta}$ neutral son todos los activos que satisfacen $\{\beta_0 S_0 + \sum \beta_n S_n = 0 \}$. Todos esos puntos me dan delta-neutral.

![](img/07.png)

**Ideal solution**  
* $P_{ideal} = L_{delta} \cap L_{beta}$ 
> la intersección de ambos hiperplanos. Si no todos los $\beta$etas son iguaes a 1, entonces los dos planos no son lineales, y por lo tanto la intersección va a ser un hiperplano de dimensión N-2
* Y todos los puntos que estén en el plano $P_{ideal}$ son beta y delta neutral. Por construcción están aquí $\{S_0 + \sum S_n = 0 \}$ y también están aquí $\{\beta_0 S_0 + \sum \beta_n S_n = 0 \}$

**EJEMPLO EN DOS DIMENSIONES:** Beta and Delta lines
    
![](img/08.png)

$L_{delta} = \{S_0 + S_1 + S_2 = 0\}$
>* Supongamos que nos dan el activo $S_0$ y tenemos dos dimensinoes desde el punto de vista de la cobertura (tengo >un activo $S_1$ y $S_2$).
>* Supongamos que tengo $S_1$==santander, $S_2$==Bbva, $S_3$==LaCaixa
>* Entonces $\{S_1 + S_2 = S_0\}$

$L_{beta} = \{\beta_0 S_0 + \beta_1 S_1 + \beta_2 S_2 = 0\}$
>* Entonces $\{\beta_1 S_1 + \beta_2 S_2 = \beta_0 S_0\}$

Suponiendo $\beta_1 > \beta_2 > \beta_3$
>* Mientras los betas no sean iguales.

Entonces tanto en delta como en beta, tenemos dos ecuaciones con dos incognitas. Por lo tanto tiene una única intersección, y ese punto es justamente el que nos va a dar una cobertura tanto delta como beta.

$P_{ideal} = L_{delta} \cap L_{beta}$
>* S_1 < 0
>* S_2 > 0

Entonces desde el punto de vista teórico, a mi me dan un activo $S_0$ inicial, yo me construyo al menos dos activos más (porque sinó no hay solución), de manera que yo encuentre cuál es la proporción de los dos activos que necesito $S_1$ y $S_2$ para tener una cobertura que sea delta-neutral y beta-neutral. Esto sería la cobertura perfecta.

Quiero un vector de unos para mis delta, el coeficiente de cada delta es uno $S_0,S_1,S_2$ --> $1,1,1$, esto es que en la matriz el reglón de deltas será puros unos.

```python
# genero matriz de dos renglones deltas, betas
mtx = np.transpose(          # traspuesta
        np.column_stack(     # acomodo como columnas
            (deltas, betas)  # relleno mis columnas
        ))
mtx
```
